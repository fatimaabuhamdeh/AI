```python
#1.Importing the libraries
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn import metrics

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#2.Importing the dataset
df = pd.read_csv("/content/Real estate.csv")
```

```python
#3.information and shape the data
df.head()
```

```python
print (df.shape)
print (df.info())
```

```python
#4.make feature and target value
X=df.drop('Y house price of unit area', axis=1)
y=df['X4 number of convenience stores']
print("X=",X.shape,"\ny=", y.shape)
```

```python
#5.Splitting the data into test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
#6.print shape the x_train and x_test
print(X_train.shape)
print(X_test.shape)
```

```python
#7.train linear regression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#8.predict from x_test (y_pred)
y_pred = model.predict(X_test)
print(y_pred)
```

```python
#9.Evaluation from metrics between y_test and y_pred
from sklearn import metrics

MAE= metrics.mean_absolute_error(y_test, y_pred)
MSE=metrics.mean_squared_error(y_test, y_pred)
RMSE= np.sqrt(MSE)

print ("Mean Absoulute Error ",MAE)
print ("Mean Squared Error ",MSE)
print ("Root Mean Squared Error ",RMSE)
```

```python
#10.correlation (seaborn)
import seaborn as sns

sns.heatmap(df.corr(), annot=True,cmap='Reds')
```
